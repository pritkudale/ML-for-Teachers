```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/pritkudale/ML-for-Teachers/refs/heads/main/Linear%20Regression/cars_24.csv')
df.head()
```

```python
# prompt: drop Unnamed: 0

df = df.drop(columns=['Unnamed: 0'])
```

```python
# prompt: use one hopt encoding for Car Name

from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit the encoder on the 'Car Name' column
encoder.fit(df[['Car Name']])

# Transform the 'Car Name' column into one-hot encoded features
encoded_features = encoder.transform(df[['Car Name']])

# Create a new DataFrame from the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(['Car Name']))

# Concatenate the encoded features with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original 'Car Name' column
df = df.drop('Car Name', axis=1)

df.head()
```

```python
df.Type.unique()
```

```python
# prompt: use OrdinalEncoder for type in order ['HatchBack' 'Sedan' 'SUV'  'Lux_sedan' 'Lux_SUV']
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder(categories=[['HatchBack', 'Sedan', 'SUV', 'Lux_sedan', 'Lux_SUV']])
df['Type_encoded'] = ordinal_encoder.fit_transform(df[['Type']])
print(df[['Type', 'Type_encoded']].head())
```

```python
df[['Type','Type_encoded']].head()
```

```python

# Target Encoding for 'Fuel'
mean_price_by_fuel = df.groupby('Fuel')['Price'].mean()
df['Fuel_target_encoded'] = df['Fuel'].map(mean_price_by_fuel)
df[['Fuel', 'Fuel_target_encoded']].head()
```

```python
print(mean_price_by_fuel)
```

```python
# Target Guided Ordinal Encoding for 'Fuel'
# Calculate the mean price for each fuel type
mean_price_by_fuel = df.groupby('Fuel')['Price'].mean().sort_values()

# Create a mapping of fuel types to their rank based on mean price
fuel_mapping = {fuel: i for i, fuel in enumerate(mean_price_by_fuel.index)}

# Apply the target guided ordinal encoding to the 'Fuel' column
df['Fuel_target_ordinal_encoded'] = df['Fuel'].map(fuel_mapping)

df[['Fuel', 'Fuel_target_ordinal_encoded']].head()
```
