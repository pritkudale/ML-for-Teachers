```python
# Import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
df=pd.read_csv('https://raw.githubusercontent.com/pritkudale/ML-for-Teachers/refs/heads/main/Linear%20Regression/Admission_Predict.csv')
df.head()
```

```python
# prompt: drop serial no

df = df.drop('Serial No.', axis=1)
df.head()
```

```python
# Rename the 'Chance of Admit ' column
df = df.rename(columns={'Chance of Admit ': 'Chance of Admit'})

# Define features (X) and target (y)
X = df.drop('Chance of Admit', axis=1)
y = df['Chance of Admit']
```

```python
# import train test split and Linear Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

print('Model test score is : ',model.score(X_test, y_test))
```

```python
from sklearn.linear_model import Lasso

# Initialize and train the lasso regression model
model = Lasso()
model.fit(X_train, y_train)

print('Model test score is : ',model.score(X_test, y_test))
```

```python
from sklearn.linear_model import Ridge

# Initialize and train the ridge regression model
model = Ridge()
model.fit(X_train, y_train)

print('Model test score is : ',model.score(X_test, y_test))
```

```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score  # Import r2_score


# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42) # Example: 5 folds

# Initialize models
linear_model = LinearRegression()
lasso_model = Lasso()
ridge_model = Ridge()


# Function to perform k-fold cross-validation
def kfold_cv(model, X, y, kf):
  scores = []
  for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    scores.append(r2_score(y_test, y_pred))  # Use r2_score instead of mean_squared_error
  return np.mean(scores), np.std(scores)

# Perform k-fold cross-validation for each model
linear_r2, linear_std = kfold_cv(linear_model, X, y, kf)
lasso_r2, lasso_std = kfold_cv(lasso_model, X, y, kf)
ridge_r2, ridge_std = kfold_cv(ridge_model, X, y, kf)


print(f"Linear Regression - Average R^2: {linear_r2:.4f}, Standard Deviation: {linear_std:.4f}")
print(f"Lasso Regression - Average R^2: {lasso_r2:.4f}, Standard Deviation: {lasso_std:.4f}")
print(f"Ridge Regression - Average R^2: {ridge_r2:.4f}, Standard Deviation: {ridge_std:.4f}")
```
